```python
# üü¢ Step 1: Install ultralytics (if not already installed)
!pip install -q ultralytics

# üü¢ Step 2: Import libraries
from ultralytics import YOLO
from google.colab import files
from google.colab.patches import cv2_imshow
import cv2

# üü¢ Step 3: Load YOLO model
model = YOLO('yolov8n.pt')   # You can use yolov8s.pt for better accuracy

# üü¢ Step 4: Upload and process image(s)
uploaded = files.upload()

for filename in uploaded.keys():
    img_path = f"/content/{filename}"

    # Run detection
    results = model(img_path)

    # Get annotated image
    annotated = results[0].plot()

    # Show result
    print(f"üñºÔ∏è Detected objects in: {filename}")
    cv2_imshow(annotated)

print("‚úÖ Detection complete!")
```
